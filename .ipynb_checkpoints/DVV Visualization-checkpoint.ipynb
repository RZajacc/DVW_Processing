### WORKPLACE

```python
def add_each_player_page(pdf, players, db, data_type='All', on_court=True, as_starter=True, as_substitute=True):

    # Loop through players dict
    for key in players:

        # Get name of player (for file to export)
        p_name = players[key]
        
        # Get players data according to specified selection type
        p_data = pl.select_players_data(p_name=p_name, ind_db=db, on_court=on_court, as_starter=as_starter, as_substitute=as_substitute)
        
        if len(p_data) > 0:
            
            # Resample data
            p_all, p_weekly, p_monthly = pl.select_and_resample_player(p_data)

            # Save fig and create page if 'All is selected'
            if data_type == 'All':

                p_all_tail = p_all.tail(15)

                # Create plot
                p_plot = pl.plot_resampled_data(p_all_tail, f'#{key} {p_name} - All games')

                # Save fig
                p_plot.savefig(f'{key} {p_name}.jpg', bbox_inches='tight')

                # Add page to report
                rep.add_page_to_report(pdf, f'{key} {p_name}.jpg', p_all_tail, resampled_data=True, cols_to_drop=['Nr', 'Name', 'Code', 'Team', 'Starter', 'Sub'])
```

```python
pdf2 = rep.PDF()
pdf2.alias_nb_pages()
```

```python
add_each_player_page(pdf2, t_players, ind,on_court=True, as_starter=True, as_substitute=True)
```

```python
pdf2.output('Test.pdf', 'F')
```
